```python
# from sklearn.linear_model import LinearRegression
# from sklearn.tree import DecisionTreeRegressor
# from sklearn.svm import SVR # this is just bad
from sklearn.ensemble import RandomForestRegressor
# from sklearn.ensemble import GradientBoostingRegressor
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import root_mean_squared_error
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score
```

## RANDOM FOREST

```python
regression = RandomForestRegressor()
```

```python
scores_regression = cross_val_score(regression, X_train, y_train,scoring="r2", cv=10)
```

```python
print("Scores:", scores_regression.max())
print("Mean:", scores_regression.mean())
print("Standard deviation:", scores_regression.std())
```

```python
regression.fit(X_train, y_train)
```

```python
y_pred_regression= regression.predict(X_test)
```

```python
R2 = r2_score(y_test , y_pred_regression)
R2
```

```python
mae = mean_absolute_error(y_test , y_pred_regression)
print("Mean Absolute Error:", mae)
```

```python
rmse = root_mean_squared_error(y_test , y_pred_regression)
rmse
```

```python
Rf_mse = mean_squared_error(y_test , y_pred_regression)
Rf_mse
```
